## 1. `clean_dataset.csv` — Credit Card Approval (690 rows, 16 columns)

**Gender** is **qualitative nominal** because even though it is encoded as 0/1, it represents a category (Male/Female) with no natural order.

**Age** is **quantitative continuous** because it is a numeric value with decimals representing a measurable age.

**Debt** is **quantitative continuous** because it represents an amount of debt — a measurable numeric value.

**Married** is **qualitative nominal** because the 0/1 encoding represents a Yes/No answer, which is a category.

**BankCustomer** is **qualitative nominal** because 0/1 represents whether someone belongs to a bank or not — a category.

**Industry** is **qualitative nominal** because it contains sector names (Industrials, Materials, etc.) with no natural order.

**Ethnicity** is **qualitative nominal** because it groups identities (White, Black, etc.) with no hierarchy.

**YearsEmployed** is **quantitative continuous** because it is a duration measured in years with decimals.

**PriorDefault** is **qualitative nominal** because 0/1 means "has previously defaulted or not" — a Yes/No category.

**Employed** is **qualitative nominal** because 0/1 indicates whether the person is employed or not — a category.

**CreditScore** is **quantitative discrete** because it is an ordered, measurable whole number score.

**DriversLicense** is **qualitative nominal** because 0/1 indicates whether someone holds a license or not — a category.

**Citizen** is **qualitative nominal** because it contains labels like "ByBirth" representing a status, with no numeric meaning.

**ZipCode** is **qualitative nominal** because even though it looks numeric, it codes a geographic location — averaging zip codes makes no sense.

**Income** is **quantitative continuous** because it is a measurable monetary value.

**Approved** is **qualitative nominal** because 0/1 represents the outcome of a decision (approved/rejected) — the categorical target variable.

## Convert Unstructured Data to Structured Data

### 1. Blog posts about travel experiences
- Method: Use text extraction + natural language processing.
- Steps:
  - Extract key fields: destination, dates, activities, accommodations, costs, ratings.
  - Apply NLP techniques: entity recognition for places, sentiment analysis for opinions, keyword extraction for themes.
  - Store results in a table or JSON schema like `{title, location, date, highlights, sentiment, tags}`.
- Reasoning: Blog posts are free-form text, so structuring requires identifying recurring travel attributes and mapping them to fields.

### 2. Audio recordings of customer service calls
- Method: Transcribe audio and then classify/extract.
- Steps:
  - Transcribe speech to text using ASR (automatic speech recognition).
  - Use NLP to identify speaker roles, issue categories, resolution status, sentiment, and key phrases.
  - Save as structured records like `{call_id, customer_name, issue_type, sentiment, duration, resolution}`.
- Reasoning: Audio itself is unstructured; transcription makes the content analyzable, then extraction creates consistent fields for reporting and analytics.

### 3. Handwritten notes from a brainstorming session
- Method: Digitize handwriting and organize by topic/action.
- Steps:
  - Scan and apply handwriting recognition (OCR for handwriting).
  - Segment notes into ideas, decisions, action items, questions.
  - Tag each item and store in a structured format like `{note_id, type, text, owner, priority}`.
- Reasoning: Handwritten notes are messy and spatially arranged; digitization plus categorization turns scattered thoughts into searchable records.

### 4. A video tutorial on cooking
- Method: Extract audio transcript and scene metadata.
- Steps:
  - Transcribe spoken instructions.
  - Detect steps, ingredients, timings, and equipment from text.
  - Optionally analyze video frames for visual cues or chapter segmentation.
  - Save as structured recipe steps: `{step_number, instruction, ingredients, duration}`.
- Reasoning: The video’s value lies in sequential instructions and ingredients, so breaking it into a step-by-step schema makes it reusable and searchable.
